```python
 #Study the provinces where Douyin bloggers are mainly located
from pyecharts.charts import Map,Geo
from pyecharts import options as opts
from pyecharts.render import make_snapshot
import pandas as pd  #pandas is a powerful data processing library
```

```python
data = pd.read_excel('美食总榜.xlsx')
data
```

```python
#Data grouping and statistics by province
data1=data.groupby('IP所属地').count()
data1
```

```python
#Take the first column of data
num=pd.Series(data1.iloc[:,0])
num
```

```python
#Output Index
num.index
```

```python
#Output the number of bloggers in the province with the most distributed bloggers
num.max()
```

```python
#Convert to list format
num=list(num)
num
```

```python
from pyecharts.render import make_snapshot
#unchained call
data=[("上海",3),("内蒙古",1),("北京",4),("四川",6),("安徽",2),("山东",1),("广东",4),("广西",1),("江苏",2),("江西",3),("山西",2),("河北",3),("河南",2),("浙江",6),("福建",5),("贵州",1),("辽宁",2),("黑龙江",2)]
map=Map()
map.add("",data,"china")
map.set_global_opts(
    title_opts=opts.TitleOpts(title="抖音美食博主影响力",subtitle="数据来源：爬虫所得数据",pos_right="center",pos_top="10%"),
    visualmap_opts=opts.VisualMapOpts(max_=12),  # Make the map colorful and set the graphic maxima
    ) 
map.render("China.html")
map.render_notebook()
```
