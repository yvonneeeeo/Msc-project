```python

from pyecharts import options as opts
from pyecharts.charts import Page, Pie
x = ['美食推荐','美食教程','美食探店']
y = [30,50,50]
c = Pie()
c.add("",[list(z) for z in zip(x, y)],radius = ["35%", "70%"])   
# Thickness and size of rings
c.set_global_opts( title_opts=opts.TitleOpts(title="Analysis"),legend_opts=opts.LegendOpts( orient="vertical", pos_top="5%", pos_left="2%" ))
c .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
c.render_notebook()
```

```python
import jieba
import numpy as np
import PIL.Image as Image
from wordcloud import WordCloud
with open("meishi.txt" ,encoding="utf-8")as file:
    #Read the text content of a file
    text=file.read()
word_list = jieba.cut(text) #generator
space_word_list = ' '.join(word_list)

# Call the open method in PIL to read the image file and generate an array via the array method in numpy.
mask_pic = np.array(Image.open("xingzhuang.jpg"))
word = WordCloud(
    font_path='C:/Windows/Fonts/方正粗黑宋简体.ttf',  # Setting fonts
    mask=mask_pic,  # Setting the background image
    background_color='#ffecec',  # Setting the background color
    max_font_size=200,  # Setting the font maximum
    max_words=3000,  # Setting the maximum number of words to be displayed
                 ).generate(space_word_list)
image = word.to_image()
word.to_file('ciyun.jpg')
image.show()
```

```python
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
import pandas as pd

df=pd.read_excel("美食教程1218_172554.xlsx") #Read file data
x = ['-肥猪猪的日常','吕厨爱美食','我是晴天','噗噗叽叽','南翔不爱吃饭''安秋金','厨房历险记','夏叔厨房'] #x-axis
y = df['粉丝数'] #y-axis

pdy = pd.Series(y)

y_value = [int(value) for value in pdy.values]  

color_js = """
            new echarts.graphic.LinearGradient(              #Gradient color generation
                                0, 
                                1, 
                                0, 
                                0,
                                [{offset: 0, color: '#ccffff'}, 
                                 {offset: 1, color: '#cccccf'}], 
                                false)
           """

def color():
    bar = Bar(init_opts=opts.InitOpts(theme='light',
                                      width='1000px',
                                      height='600px'))
    bar.add_xaxis(x) #x-axis
    bar.add_yaxis("粉丝数(w)", y_value,category_gap=25,  #y-axis
                  # Executing gradient color code with JsCode
                  itemstyle_opts=opts.ItemStyleOpts(color=JsCode(color_js)))
    bar.reversal_axis() #orthogonal
    
    return bar


chart =color()
chart.render_notebook()





```
